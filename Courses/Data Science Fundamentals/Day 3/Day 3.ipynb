### Recap of Last Class

```python
import numpy as np

# Random numbers creation
arr = np.random.randint(0,100, (4,5))
print(arr)
```

```python
# Indexing and Slicing: 1) Get the second item (sub-array) in the main array
print("Example 1")
print(arr[1])

# 2) Get the last two items in the second and third sub-arrays
print("\nExample 2")
print(arr[1:3, 3:])

# 3) Get the first and last items of all the sub-arrays
print("\nExample 3")
print(arr[:, [0, -1]])
```

```python
# Universal functions: Functions that apply to every item in the array in an operation called broadcasting
# Examples are np.log, np.add, np.multiply, np.sqrt, np.power, etc
print("Example 1")
print(np.exp(arr))
print("\nExample 2")
print(np.add(arr, arr))
print("\nExample 3")
print(np.square(arr))
```

```python
# 3: Some Array attributes and methods: `dtype`, `reshape`, `shape`, `flatten`, etc
print("Example 1")
print(arr.dtype) # Attribute

print("\nExample ")
print(arr.shape) # Attribute

print("\nExample 3")
print(arr.reshape(2,10)) # method

print("Example 1")
print(arr.flatten()) # method
```

```python
print(arr)
print()
print(arr.reshape(2,10))
```

### Selecting a Subset of a Dataframe

#### Columns Subsetting

```python
# Example 1: Select Number of deaths
no_of_deaths = covid_df['No. of Deaths']
type(no_of_deaths)
```

```python
no_of_deaths = covid_df[['No. of Deaths']]
no_of_deaths
```

```python
# Example 2: Select No. of Cases (Lab Confirmed) and No. Discharged columns

covid_df[[ 'No. of Deaths', 'No. of Cases (Lab Confirmed)']].head()
```

```python
covid_df[[ 'No. of Deaths', 'No. of Cases (Lab Confirmed)']].shape
```

#### Row Subsetting

```python
# Creating a mask: A mask is a series or dataframe of True and False values
# For instance, let us create a mask of records with more than 30 deaths
mask = covid_df['No. of Deaths']>30
mask
```

**When the mask is used to filter the dataframe, only the values that are true are returned**

```python
# Example 1: Selecting the records of states with more than 15 deaths
mask = covid_df['No. of Deaths']>15
covid_df[mask]
```

```python
# Example 2: Selecting states with more than 250 confirmed cases and less than 500 cases
mask = (covid_df['No. of Cases (Lab Confirmed)']>250) & (covid_df['No. of Cases (Lab Confirmed)']<500)
covid_df[mask]
```

```python
# Example 3: Getting the records of FCT and Lagos
mask = (covid_df['States Affected'] == 'FCT')|(covid_df['States Affected'] == 'Lagos')
covid_df[mask]
```

### Creating Plots with Pandas

**Pandas plot function is an extension of Matplotlib library.**

For this, we will be using the data on air quality in this [link](https://github.com/pandas-dev/pandas/tree/master/doc/data/air_quality_no2.csv)

```python
# Read the data

df = pd.read_html('https://github.com/pandas-dev/pandas/tree/master/doc/data/air_quality_no2.csv',
                  index_col=1)[0].drop('Unnamed: 0', 1)
```

Inspect the data

```python
df.head()
```

Create a timeseries plot

```python
df.plot(figsize = (12, 10))
```

Plotting for just London station

```python
df['station_london'].plot(figsize = (16, 12))
```

**Scatter plot on London vs Paris air qualities**

```python
df
```

```python
df.plot(x = 'station_london', y = 'station_paris', kind = 'scatter', figsize = (16,9))
```

**Let us try boxplots**

```python
df.plot(kind = 'box', figsize = (16, 12))
```

**You can also put each column in a separate plot**

```python
ax = df.plot(kind = 'hist', figsize = (6, 16), subplots = True, bins =20)
```

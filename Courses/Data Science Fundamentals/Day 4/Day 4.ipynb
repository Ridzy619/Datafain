## Creating new columns derived from existing columns; renaming columns

Using the California housing dataset referenced earlier with a link:

```python
import pandas as pd

# Example: Read a csv file
file = r"C:\Users\HP\Documents\Ridwan\DataFAIN\Datafain\Courses\Data Science Fundamentals\Day 4\california_housing_train.csv"
df = pd.read_csv(file)
df.head()
```

**Let us create a new column which is the number of people per household**

```python
df['population_per_household'] = df['population'] / df['households']
df.head()
```

**Let us rename `median_house_value` and `population_per_household` to `med_house_val` and `pop_per_house` respectively**

**Let us also rename indices `0` and `1` to `00` and `11` respectively**

```python
col_new_names = {'median_house_value':'med_house_val',
                 'population_per_household': 'pop_per_house'}
index_new_names = {0:'00',
                1:'11'}

df.rename(columns=col_new_names, index= index_new_names)
```

## Combining DataFrames

Dataframes can be combined with concat, merge or join. All three methods do similar things, but they can also be more suited to certain tasks than others.

For example, concat method is more suited to combining data along row or column axes with the assumption that the records that are being combined are aligned.

The focus in this class will be the merge method that is used to combine two tables by specifying the primary column to merge on.

```python
import numpy as np
```

```python
df1 = pd.DataFrame(data= {'A':np.random.randint(0,100, (10,)),
                    'B':np.random.rand(10)},
             index = range(10))
df2 = pd.DataFrame(data= {'A':np.random.randint(0,100, (10,)),
                    'B':np.random.rand(10)},
             index = range(10))
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

## Handling missing data
Missing data are usually represented as `NaN` in Pandas. Such values can pose problem when we are trying to build machine learning models or analyse data. For this reason, they have to be handled before proceeding with processing.

Pandas has built-in methods for handling missing data such as `fillna` and `dropna`. We will examine how to use these two methods.

```python
# df.fillna
df = pd.read_html('https://github.com/pandas-dev/pandas/blob/master/doc/data/air_quality_no2.csv')[0].drop('Unnamed: 0', axis = 1)
df
```

**`fillna`**

This method has a number of ways to fillna, such as foward fill using the value of the next record, backward fill using the previous record, etc.

```python
# Example 1: Fill nan values with a specified value
df.fillna(0)
```

```python
# Example 2: Forward fill
df.fillna(method = 'ffill')
```

```python
# Example 3: Backward fill
df.fillna(method = 'bfill')

# Alternatively, we can do:
df.bfill()
```

```python
# Example 3: Fix the errors from above by using both fill methods
df.ffill().bfill()
```

### Using the Object oriented interface

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# You can create multiple plots like this

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'b*', t2, f(t2), 'k')


plt.subplot(212)
plt.plot(t2, np.cos(np.pi*t2), 'r--')
plt.show()
```

### Creating multiple plots

```python
x = np.arange(10)

# Creating a figure and specifying it's axes
fig, axs = plt.subplots(2,2, figsize=(16,16))

# Unpack the axes array created
(ax1, ax2), (ax3, ax4)= axs

ax1.plot(x, np.random.randn(10)*x)
ax2.scatter(x*np.random.randn(10), x**2)
ax3.bar(x, x**3/2)
ax4.pie(x, labels = np.arange(10), autopct ='%1.2f%%')
plt.show()
```

### More on plot formatting

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot([1,2, 3, 5], [3,6, 20, 50], "bo", [2,5, 10, 25], [5,25, 125, 60], "r*")
ax.text(12, 50, "Text illustration", fontsize = 12, color="g")
ax.set_title("Hello World")
ax.set_label("Blue")
plt.grid(lw = 1.25)

plt.show()
```
